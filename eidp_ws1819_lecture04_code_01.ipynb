## Example 04.01 - More Than One Circle

```java
float positionX1 = 0;
float positionY1 = 0;
float positionX2 = 0;
float positionY2 = 0;
float positionX3 = 0;
float positionY3 = 0;

float stepX1;
float stepY1;
float stepX2;
float stepY2;
float stepX3;
float stepY3;

int hue1 = 0;
int hue2 = 360;
int hue3 = 180;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    stepX1 = random(-5, 5);
    stepY1 = random(-5, 5);
    stepX2 = random(-5, 5);
    stepY2 = random(-5, 5);
    stepX3 = random(-5, 5);
    stepY3 = random(-5, 5);
}


void draw()
{
    background(0, 0, 100);
    
     //1
    if(positionX1 < 0 || positionX1 >= width)
    {
        stepX1 *= -1;
    }
    
    if(positionY1 < 0 || positionY1 >= height)
    {
        stepY1 *= -1;
    }
    
    positionX1 += stepX1;
    positionY1 += stepY1;
    
    hue1++;
    if(hue1 >= 360)
    {
        hue1 = 0;
    }
    fill(hue1, 100, 100);
    ellipse(positionX1, positionY1, 10, 10);
    
    //2
    if(positionX2 < 0 || positionX2 >= width)
    {
        stepX2 *= -1;
    }
    
    if(positionY2 < 0 || positionY2 >= height)
    {
        stepY2 *= -1;
    }
    
    positionX2 += stepX2;
    positionY2 += stepY2;
    
    hue2--;
    if(hue2 >= 360)
    {
        hue2 = 0;
    }
    fill(hue2, 100, 100);
    ellipse(positionX2, positionY2, 10, 10);
    
    //3
    if(positionX3 < 0 || positionX3 >= width)
    {
        stepX3 *= -1;
    }
    
    if(positionY3 < 0 || positionY3 >= height)
    {
        stepY3 *= -1;
    }
    
    positionX3 += stepX3;
    positionY3 += stepY3;
    
    hue3++;
    if(hue3 >= 360)
    {
        hue3 = 0;
    }
    fill(hue3, 100, 100);
    ellipse(positionX3, positionY3, 10, 10);
}



```

## Example 04.02 - While Loop

```java
int counter = 0;    // Iteration counter

while (counter < 3) // Loop condition
{  
    println("wow");  // Loop activity
    counter += 1;   // Increase counter
}
println("All done…");
```

This is how you create an infinity loop. Avoid this at all times!

Solution:

```java
int numCircles = 30;

float[] positionX = new float[numCircles];
float[] positionY = new float[numCircles];

float[] stepX = new float[numCircles];
float[] stepY = new float[numCircles];

float[] hue = new float[numCircles];
float[] radius = new float[numCircles];

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    
    int counter = 0;
    while(counter < numCircles)
    {
        positionX[counter] = random(width);
        positionY[counter] = random(height);
        
        stepX[counter] = random(-5, 5);
        stepY[counter] = random(-5, 5);
        
        hue[counter] = random(360);
        radius[counter] = random(5,15);
        
        counter++;
    }
}


void draw()
{
    background(0, 0, 100);
    int counter = 0;
    
    while(counter < numCircles)
    {
        
        if(positionX[counter] < 0 || positionX[counter] >= width)
        {
            stepX[counter] *= -1;
        }

        if(positionY[counter] < 0 || positionY[counter] >= height)
        {
            stepY[counter] *= -1;
        }

        positionX[counter] += stepX[counter];
        positionY[counter] += stepY[counter];

        

        fill(hue[counter], 100, 100);

        ellipse(positionX[counter], positionY[counter], radius[counter], radius[counter]);
        
        counter++;
    }
}



// Called if a key was 
// pressed
void keyPressed()
{
    if(key == 'c')
    {
        background(0, 0, 100);
    }
}
```

With For Loops:

```java
int numCircles = 30;

float[] positionX = new float[numCircles];
float[] positionY = new float[numCircles];

float[] stepX = new float[numCircles];
float[] stepY = new float[numCircles];

float[] hue = new float[numCircles];
float[] radius = new float[numCircles];

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    
    for(int i = 0; i < numCircles; i++)
    {
        positionX[i] = random(width);
        positionY[i] = random(height);
        
        stepX[i] = random(-5, 5);
        stepY[i] = random(-5, 5);
        
        hue[i] = random(360);
        radius[i] = random(5,15);
    }
}


void draw()
{
    background(0, 0, 100);
    
    for(int i = 0; i < numCircles; i++)
    {
        
        if(positionX[i] < 0 || positionX[i] >= width)
        {
            stepX[i] *= -1;
        }

        if(positionY[i] < 0 || positionY[i] >= height)
        {
            stepY[i] *= -1;
        }

        positionX[i] += stepX[i];
        positionY[i] += stepY[i];

        

        fill(hue[i], 100, 100);

        ellipse(positionX[i], positionY[i], radius[i], radius[i]);
    }
}



// Called if a key was 
// pressed
void keyPressed()
{
    if(key == 'c')
    {
        background(0, 0, 100);
    }
}
```

## Example 04.08 - For Loop Examples

Local scope:

```java
for(int i = 0; i < 10; i++)
{
    println("i inside the loop: " + i); //gives an error
}
//println("i outside the look: " + i); //gives an error
```

Accessing variables from a higher scope:

```java
int sum = 0;

for(int i = 0; i < 10; i++)
{
    println("i inside the loop: " + i); //gives an error
    sum++;
}
println("Sum: " + sum);
```

Counting backwards:

```java
for(int i = 10; i >= 0; i--)
{
    println("i inside the loop: " + i); //gives an error
}
```

Every second number as iterator:

```java
for(int i = 0; i < 10; i+=2)
{
    println("i inside the loop: " + i); //gives an error
}
```
