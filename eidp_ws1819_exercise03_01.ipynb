### Task 03.01

Take the code example 03.09 from the lecture as basis. Replace the mouse following of the circle with moving the circle with the arrow keys of the keyboard (up, down, left, right).

Use Processing's keyCode and read the reference on how to use it.

Hint
- The controll and adjustment of the position variables happens now in the keyPressed() function.

```java
float positionX = 0;
float positionY = 0;

float speed;

int hue = 0;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    speed = 5;
}


void draw()
{
    
    hue++;
    if(hue >= 360)
    {
        hue = 0;
    }
    fill(hue, 100, 100);
    ellipse(positionX, positionY, 10, 10);
}



// Called if a key was 
// pressed
void keyPressed()
{
    
    if (key == CODED) 
    {
        if (keyCode == UP) 
        {
            positionY -= speed;
        } 
        else if (keyCode == DOWN) 
        {
            positionY += speed;
        }
        else if (keyCode == LEFT) 
        {
            positionX -= speed;
        } 
        else if (keyCode == RIGHT) 
        {
            positionX += speed;
        } 
    } 
    else 
    {
        // Clearing the screen
        if(key == 'c')
        {
            background(0, 0, 100);
        }
    }
}
```

### Task 03.03

Adjust your program so that the collision with the sktech window boarders is checked with the outside of the circle, not its center. The inital start position of the circle should be in the center of the sktech window.

```java
float positionX;
float positionY;

float speed;

int radius;

int hue = 0;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    positionX = 0.5 * width;
    positionY = 0.5 * height;
    radius = 40;
    speed = 5;
}


void draw()
{
    
    hue++;
    if(hue >= 360)
    {
        hue = 0;
    }
    fill(hue, 100, 100);
    ellipse(positionX, positionY, 2*radius, 2*radius);
}



// Called if a key was 
// pressed
void keyPressed()
{
    
    if (key == CODED) 
    {
        if (keyCode == UP) 
        {
            if(positionY >= radius)
            {
                positionY -= speed;
            }
        } 
        else if (keyCode == DOWN) 
        {
            if(positionY <= (height - radius))
            {
                positionY += speed;
            }  
        }
        else if (keyCode == LEFT) 
        {
            if(positionX >= radius)
            {
                positionX -= speed;
            } 
            
        } 
        else if (keyCode == RIGHT) 
        {
            if(positionX <= (width - radius))
            {
                positionX += speed;
            } 
            
        } 
    } 
    else 
    {
        // Clearing the screen
        if(key == 'c')
        {
            background(0, 0, 100);
        }
    }
}
```
