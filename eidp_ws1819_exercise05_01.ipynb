# Task 05.01
    
Add an initCoins() function.

```java
// AVATAR
float positionX;
float positionY;

float speed;

int radius;
int hue = 0;

//COINS
int coinsSize = 15;
int coinsNum = 16;
float[] coinsPositionX = new float[coinsNum];
float[] coinsPositionY = new float[coinsNum];
boolean[] coinsActive = new boolean[coinsNum];


void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 0);
    fill(200, 100, 100);
    noStroke();
    
    positionX = 0.5 * width;
    positionY = 0.5 * height;
    radius = 15;
    speed = 5;
    
    
    // 05.01 Re-structuring the code
    initCoins();
}


void draw()
{
    background(0, 0, 0);
    
    updateCoins();
    updateAvatar();   
}



void updateAvatar()
{
        // AVATAR
    hue++;
    if(hue >= 360)
    {
        hue = 0;
    }
    fill(hue, 100, 100);
    ellipse(positionX, positionY, 2*radius, 2*radius);
    
    
}


void initCoins()
{
    for(int i = 0; i < coinsNum; i++)
    {
        coinsPositionX[i] = random(coinsSize, width - coinsSize);
        coinsPositionY[i] = random(coinsSize, height - coinsSize);  
        coinsActive[i] = true;
    } 
}


void updateCoins()
{
    // COINS
    for(int i = 0; i < coinsNum; i++)
    {
        
        if(coinsActive[i] == true)
        {
            
            //Check for a collision of the avatar with a coin
            if(positionX >= coinsPositionX[i] && positionX <= coinsPositionX[i] + coinsSize &&
                positionY >= coinsPositionY[i] && positionY <= coinsPositionY[i] + coinsSize)
            {
                coinsActive[i] = false;
                println("Yummy!");
            }
            else
            {
                fill(40, 100, 100);
                rect(coinsPositionX[i], coinsPositionY[i], coinsSize, coinsSize);
            }
        }
    }
}



// Called if a key was 
// pressed
void keyPressed()
{
    
    if (key == CODED) 
    {
        if (keyCode == UP) 
        {
            if(positionY >= radius)
            {
                positionY -= speed;
            }
        } 
        else if (keyCode == DOWN) 
        {
            if(positionY <= (height - radius))
            {
                positionY += speed;
            }  
        }
        else if (keyCode == LEFT) 
        {
            if(positionX >= radius)
            {
                positionX -= speed;
            } 
            
        } 
        else if (keyCode == RIGHT) 
        {
            if(positionX <= (width - radius))
            {
                positionX += speed;
            } 
            
        } 
    } 
    else 
    {
        // Clearing the screen
        if(key == 'c')
        {
            background(0, 0, 100);
        }
    }
}
```

# Task 05.02 (2 Points)
    
Distribute the coins evenly on a grid.

There a several ways to solve this task. One way is to follow example 04.02 from lecture as basis. For a better exercise, first try to come to a solution without reading the hints.

Hints to complete the tasks:
- As we are drawing reaguarly on a grid now, we don't need to save the coins positions anymore as we did it with the random values. We can just re-compute the positions in each draw call. Hence we do not need the variables coinsPositionX and coinsPositionY anymore - delete thouse variables.
- We divide the canvas into tiles and draw the coins at the center of each tile (example 04.02 from lecture):
    - Create the global integer variables tileWidth and tileHeight
    - In the initCoins() function compute the width of one tile by deviding the canvas width by the number of coins. Do the same for the height of a tile.
    - In the updateCoins() function iterate through all rows and coluoms and use tileHeight and tileWidth as step size for the loop.


```java
// AVATAR
float positionX;
float positionY;

float speed;

int radius;
int hue = 0;

//COINS
int coinsSize = 15;
int coinsNum = 16;

boolean[] coinsActive = new boolean[coinsNum];


// 05.02 ADD VARIABLES FOR THE TILE SIZE
int tileWidth;
int tileHeight;


void setup()
{
    size(512, 512); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 0);
    fill(200, 100, 100);
    noStroke();
    
    positionX = 0.5 * width;
    positionY = 0.5 * height;
    radius = 15;
    speed = 5;
    
    
    initCoins();
}


void draw()
{
    background(0, 0, 0);
    
    updateCoins();
    updateAvatar();   
}



void updateAvatar()
{
    // AVATAR
    hue++;
    if(hue >= 360)
    {
        hue = 0;
    }
    fill(hue, 100, 100);
    ellipse(positionX, positionY, 2*radius, 2*radius);
    
    
}


void initCoins()
{
    // 05.02 COMPUTE THE TILE WIDTH AND HEIGHT
    tileWidth = (width - 1) / coinsNum;
    tileHeight = (height - 1) / coinsNum;
    
    for(int i = 0; i < coinsNum; i++)
    {
        // 05.02 REMOVE RANDOM POSITIONS 
        coinsActive[i] = true;
    } 
}


void updateCoins()
{
    // COINS
    
    // 05.02 ADD 2D LOOP FOR PLACING COINS EVENLY
    for (int y = 0; y < height; y += tileHeight) 
    {
        for (int x = 0; x < width; x += tileWidth) 
        {
        
            fill(40, 100, 100);
            ellipse(x, y, coinsSize, coinsSize);
        }
    }
}



// Called if a key was 
// pressed
void keyPressed()
{
    
    if (key == CODED) 
    {
        if (keyCode == UP) 
        {
            if(positionY >= radius)
            {
                positionY -= speed;
            }
        } 
        else if (keyCode == DOWN) 
        {
            if(positionY <= (height - radius))
            {
                positionY += speed;
            }  
        }
        else if (keyCode == LEFT) 
        {
            if(positionX >= radius)
            {
                positionX -= speed;
            } 
            
        } 
        else if (keyCode == RIGHT) 
        {
            if(positionX <= (width - radius))
            {
                positionX += speed;
            } 
            
        } 
    } 
    else 
    {
        // Clearing the screen
        if(key == 'c')
        {
            background(0, 0, 100);
        }
    }
}
```
