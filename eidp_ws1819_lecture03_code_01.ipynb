## Example 03.01 - Variable

```java
// Called once when
// the program is started
void setup()
{
    size(250, 250); 
    background(255); 
    
    noStroke();    
}

// Called continuesly
// until the program
// is quit
void draw()
{

    float distance = abs(mouseX - pmouseX) + abs(mouseX - pmouseX);
    ellipse(mouseX, mouseY, distance, distance); 
}

// Called if a key was 
// pressed
void keyPressed()
{
    if(key == 'r')
    {
        fill(random(100, 220), 0, 0);
    }
    
    if(key == 'g')
    {
        fill(0, random(100, 220), 0);
    }
    
    if(key == 'b')
    {
        fill(0, 0, random(100, 220));
    }
    
    
    if(key == 'c')
    {
        background(255);
    }
    
    
}
```

## Example 03.03 - Variable Scope

```java
float floatingPointNumber = 10.25;

void setup()
{
    size(250, 250); 
    background(255);
    
    noLoop();
    
    int wholeNumber = 10;
}


void draw()
{
    boolean truthValue = false;
    String word = "Hello World";
    
    
    // This will throw an error as is
    // because the variable wholeNumber
    // is defined in setup() which
    // is a different scope
    println(wholeNumber); 
    
    
    println(floatingPointNumber);
    println(truthValue);
    println(word);
}
```

## Example 03.04 - Nested Variables

```java
void setup()
{
    noLoop();
}


void draw()
{
    float floatingPointNumber = 10.25;
    
    if(floatingPointNumber == 10.25)
    {
        println(floatingPointNumber);
        int wholeNumber = 10;
    }
    
    // This will throw an error
    println(wholeNumber);

}
```

Global Scope:

```java
float floatingPointNumber = 10.25;

void setup()
{
    noLoop();
}


void draw()
{
    println(floatingPointNumber);
}
```

```java
void setup()
{
    size(250, 250); 
    background(255);
    
    noLoop();
    
    int wholeNumber = 10;
    println(wholeNumber);
}


void draw()
{
    int wholeNumber = 20;
    println(wholeNumber);
}
```

## Example 03.05 - Animating The Circle

```java
int positionX = 0;
int positionY = 0;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
}


void draw()
{
    positionX += 5;
    positionY += 5;
    ellipse(positionX, positionY, 10, 10);
}



// Called if a key was 
// pressed
void keyPressed()
{
    // Set the fill color to a 
    // randomly chosen one
    fill(random(360), 100, 100);

    if(key == 'c')
    {
        background(0, 0, 100);
    }
}
```

## Example 03.07 - Changing Direction

```java
int positionX = 0;
int positionY = 0;

int stepX = 5;
int stepY = 5;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
}


void draw()
{
 
    if(positionX < 0 || positionX >= width)
    {
        println("x is outside the window", positionX);
        stepX *= -1;
    }
    
    if(positionY < 0 || positionY >= height)
    {
        println("y is outside the window", positionY);
        stepY *= -1;
    }
    
    positionX += stepX;
    positionY += stepY;
    
    ellipse(positionX, positionY, 10, 10);
}



// Called if a key was 
// pressed
void keyPressed()
{
    // Set the fill color to a 
    // randomly chosen one
    fill(random(360), 100, 100);

    if(key == 'c')
    {
        background(0, 0, 100);
    }
}
```

## Example 03.08 - Randomizing the Step Size

```java
int positionX = 0;
int positionY = 0;

int stepX;
int stepY;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    stepX = random(-5, 5);
    stepY = random(-5, 5);
}


void draw()
{
 
    if(positionX < 0 || positionX >= width)
    {
        println("x is outside the window", positionX);
        stepX *= -1;
    }
    
    if(positionY < 0 || positionY >= height)
    {
        println("y is outside the window", positionY);
        stepY *= -1;
    }
    
    positionX += stepX;
    positionY += stepY;
    
    ellipse(positionX, positionY, 10, 10);
}



// Called if a key was 
// pressed
void keyPressed()
{
    // Set the fill color to a 
    // randomly chosen one
    fill(random(360), 100, 100);

    if(key == 'c')
    {
        background(0, 0, 100);
    }
}
```

## Example 09 - Animation of the Color

```java
float positionX = 0;
float positionY = 0;

float stepX;
float stepY;

int hue = 0;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    stepX = random(-5, 5);
    stepY = random(-5, 5);
}


void draw()
{
 
    if(positionX < 0 || positionX >= width)
    {
        stepX *= -1;
    }
    
    if(positionY < 0 || positionY >= height)
    {
        stepY *= -1;
    }
    
    positionX += stepX;
    positionY += stepY;
    
    hue++;
    fill(hue, 100, 100);

    ellipse(positionX, positionY, 10, 10);
}



// Called if a key was 
// pressed
void keyPressed()
{
    if(key == 'c')
    {
        background(0, 0, 100);
    }
}
```

Setting hue back to 0:

```java
float positionX = 0;
float positionY = 0;

float stepX;
float stepY;

int hue = 0;

void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    stepX = random(-5, 5);
    stepY = random(-5, 5);
}


void draw()
{
 
    if(positionX < 0 || positionX >= width)
    {
        stepX *= -1;
    }
    
    if(positionY < 0 || positionY >= height)
    {
        stepY *= -1;
    }
    
    positionX += stepX;
    positionY += stepY;
    
    hue++;
    if(hue >= 360)
    {
        hue = 0;
    }
    fill(hue, 100, 100);
    ellipse(positionX, positionY, 10, 10);
}



// Called if a key was 
// pressed
void keyPressed()
{
    if(key == 'c')
    {
        background(0, 0, 100);
    }
}
```
