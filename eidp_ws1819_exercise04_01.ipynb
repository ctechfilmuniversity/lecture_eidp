Storing values in  the variables and arrays 
* In the setup function create a loop with as many iterations as number of coins. In each iteration, assign each x and y array position a random number within the canvas (remember: width, height).

Drawing the coins
* In the draw function create a loop with as many iterations as number of coins. In each iteration draw a rectangle of coin size at the positions that are stored in the position arrays.

### Bonus

With an `active` property for the coins, saved in another array, we control the drawing of the coins. We only want to draw coins which weren't eaten yet.

```java
// AVATAR
float positionX;
float positionY;

float speed;

int radius;
int hue = 0;

//COINS
int coinsSize = 15;
int coinsNum = 10;
float[] coinsPositionX = new float[coinsNum];
float[] coinsPositionY = new float[coinsNum];
boolean[] coinsActive = new boolean[coinsNum];


void setup()
{
    size(250, 250); 
    
    colorMode(HSB, 360, 100, 100);
    background(0, 0, 100);
    fill(200, 100, 100);
    noStroke();
    
    positionX = 0.5 * width;
    positionY = 0.5 * height;
    radius = 15;
    speed = 5;
    
    
    for(int i = 0; i < coinsNum; i++)
    {
        coinsPositionX[i] = random(coinsSize, width - coinsSize);
        coinsPositionY[i] = random(coinsSize, height - coinsSize);  
        coinsActive[i] = true;
    }
}


void draw()
{
    background(0, 0, 100);
    
    updateCoins();
    updateAvatar();   
}



void updateAvatar()
{
        // AVATAR
    hue++;
    if(hue >= 360)
    {
        hue = 0;
    }
    fill(hue, 100, 100);
    ellipse(positionX, positionY, 2*radius, 2*radius);
    
    
}

void updateCoins()
{
    // COINS
    for(int i = 0; i < coinsNum; i++)
    {
        
        if(coinsActive[i] == true)
        {
            
            //Check for a collision of the avatar with a coin
            if(positionX >= coinsPositionX[i] && positionX <= coinsPositionX[i] + coinsSize &&
                positionY >= coinsPositionY[i] && positionY <= coinsPositionY[i] + coinsSize)
            {
                coinsActive[i] = false;
                println("Yummy!");
            }
            else
            {
                fill(40, 100, 100);
                rect(coinsPositionX[i], coinsPositionY[i], coinsSize, coinsSize);
            }
        }
    }
}



// Called if a key was 
// pressed
void keyPressed()
{
    
    if (key == CODED) 
    {
        if (keyCode == UP) 
        {
            if(positionY >= radius)
            {
                positionY -= speed;
            }
        } 
        else if (keyCode == DOWN) 
        {
            if(positionY <= (height - radius))
            {
                positionY += speed;
            }  
        }
        else if (keyCode == LEFT) 
        {
            if(positionX >= radius)
            {
                positionX -= speed;
            } 
            
        } 
        else if (keyCode == RIGHT) 
        {
            if(positionX <= (width - radius))
            {
                positionX += speed;
            } 
            
        } 
    } 
    else 
    {
        // Clearing the screen
        if(key == 'c')
        {
            background(0, 0, 100);
        }
    }
}
```
